```python
#importing the libraries 
import pandas as pd 
import seaborn as sns 
import numpy as np 

import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

#read the data 
df = pd.read_csv (r'C:\Users\Hp\Desktop\movies.csv')
# looking onto the data
df.head()
# check any missing data 
import numpy as np

# Assuming 'df' is the DataFrame you want to analyze

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) * 100
    print(f"The percentage of missing values in column '{col}' is: {pct_missing}%")
#data types
df.dtypes
```

```python
df
```

```python
mean_budget = df['budget'].mean()
df['budget'] = df['budget'].fillna(mean_budget)
```

```python
df
```

```python
df = df.dropna(subset=['gross'])
```

```python
df
```

```python
#budget and gross are correlated 
```

```python
plt.scatter(df['budget'], df['gross'])
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.title('Relationship between Budget and Gross')
plt.show()
```

```python

# Assuming 'df' is the DataFrame containing the 'budget' and 'gross' columns

# Define colors for scatter points and regression line
scatter_color = "#FF6F61"
line_color = "#6B5B95"

# Create the scatter plot with specified colors
sns.regplot(data=df, x='budget', y='gross', scatter_kws={'color': scatter_color}, line_kws={'color': line_color})

# Set the title for the scatter plot
plt.title('Scatter Plot: Budget vs Gross')

# Display the plot
plt.show()
```

```python
df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
```

```python
df_numerized = df.copy()  # Make a copy of the original DataFrame

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == object:
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

# Display the numerized DataFrame
print(df_numerized)
```
